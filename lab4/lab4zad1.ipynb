```julia
function f(x)
    return 1/(1+25*x*x)
end

function LHelper(i, x, N, Array, ValuesArray, LHelperArr)
    result =1
    for j=1:N
        if j!=i
            result*= (x-Array[j])/(Array[i]-Array[j])
        end
    end
    push!(LHelperArr, result)
    return result
end

function Lagrange(x, N, Array, ValuesArray, LHelperArr)
    result=0
    for i=1:N
        result+= ValuesArray[i]*LHelper(i, x, N, Array, ValuesArray, LHelperArr)
    end
    return result
end
```

```julia
f(0.2)
```

```julia
PyPlot.plot(FunctionArrayClose, hcat(FunctionValuesClose, LagrangeValuesClose))
```
